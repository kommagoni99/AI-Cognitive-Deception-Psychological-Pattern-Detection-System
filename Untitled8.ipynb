```python
# ==========================================================
# AI COGNITIVE DECEPTION & PSYCHOLOGICAL PATTERN DETECTOR
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
import textstat
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to resolve LookupError for punkt_tab

np.random.seed(42)

# ----------------------------------------------------------
# 1. SYNTHETIC DATA GENERATION (Truth vs Deception)
# ----------------------------------------------------------

truth_samples = [
    "I went to the office at 9 AM and completed my work.",
    "I had lunch with my colleague at 1 PM.",
    "I attended the meeting and submitted the report.",
    "I visited my parents last weekend.",
    "I studied for my exams in the evening."
]

deception_samples = [
    "Honestly I swear I was absolutely there and completely finished everything perfectly.",
    "To be very honest I did not do anything wrong at all.",
    "I promise I totally was not involved in anything suspicious.",
    "Believe me I definitely did not forget anything important.",
    "Trust me I absolutely completed everything flawlessly."
]

data = []
for _ in range(300):
    data.append([np.random.choice(truth_samples), 0])
    data.append([np.random.choice(deception_samples), 1])

df = pd.DataFrame(data, columns=["text", "label"])

# ----------------------------------------------------------
# 2. FEATURE ENGINEERING (Psychological Indicators)
# ----------------------------------------------------------

def extract_features(text):
    words = nltk.word_tokenize(text)
    word_count = len(words)
    avg_word_len = np.mean([len(w) for w in words])
    exclamation_count = text.count("!")
    readability = textstat.flesch_reading_ease(text)
    pronouns = len(re.findall(r'\b(I|me|my|mine)\b', text.lower()))
    emotion_words = len(re.findall(r'\b(honestly|swear|promise|trust|definitely|absolutely)\b', text.lower()))

    return [
        word_count,
        avg_word_len,
        exclamation_count,
        readability,
        pronouns,
        emotion_words
    ]

features = np.array([extract_features(t) for t in df["text"]])
labels = df["label"]

# ----------------------------------------------------------
# 3. TRAIN TEST SPLIT
# ----------------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)

# ----------------------------------------------------------
# 4. MODEL TRAINING
# ----------------------------------------------------------

model = RandomForestClassifier(n_estimators=150)
model.fit(X_train, y_train)

# ----------------------------------------------------------
# 5. EVALUATION
# ----------------------------------------------------------

preds = model.predict(X_test)

print("\nClassification Report:\n")
print(classification_report(y_test, preds))

cm = confusion_matrix(y_test, preds)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap="Reds")
plt.title("Deception Detection Confusion Matrix")
plt.show()

# ----------------------------------------------------------
# 6. FEATURE IMPORTANCE VISUALIZATION
# ----------------------------------------------------------

importance = pd.Series(model.feature_importances_,
                       index=["WordCount","AvgWordLen","Exclamations","Readability","Pronouns","EmotionWords"])

plt.figure(figsize=(8,5))
importance.sort_values().plot(kind="barh")
plt.title("Psychological Feature Importance")
plt.show()

# ----------------------------------------------------------
# 7. LIVE DECEPTION ANALYZER
# ----------------------------------------------------------

print("\n------ LIVE COGNITIVE DECEPTION ANALYZER ------")
user_input = input("Enter a statement: ")

user_features = np.array(extract_features(user_input)).reshape(1, -1)

prediction = model.predict(user_features)[0]
prob = model.predict_proba(user_features)[0][1]

if prediction == 1:
    print("\n⚠ Potential Deception Detected")
else:
    print("\n✅ Statement Appears Truthful")

print("Deception Probability:", round(prob*100,2), "%")
```
